```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
pd.set_option('display.max_columns',None)
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names=final_df['Country'].value_counts().index
```

```python
country_val=final_df['Country'].value_counts().values
```

```python
# pie chart - top 3 countries uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation : Zomato maximum records/transaction were in india,then United States and then its United Kingdom

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observations
1. Rating is between 1.8 to 2.4 ---> Poor
2. Rating is between 2.5 to 3.4 ---> Average
3. Rating is between 3.5 to 3.9 ---> Good
4. Rating is between 4.0 to 4.4 ---> Very Good
5. Rating is between 4.5 to 4.9 ---> Excellent

```python
ratings.head()
```

```python

plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

## Observations:
1. Not rated count is very hight
2. Maximum no of ratings are between 2.5 and 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
